## 16. and, or 그리고 not 연산자
다음 코드를 실행한 결과를 적어 보세요

````python
a = 20
b = 30
print(a > 40 and b > 40)
print(a > 20 and b > 20)
print(a < 30 and b > 30)
print(a > 10 and b > 10)
print(20 > 40 or 30 > 40)
print(20 > 20 or 30 > 20)
print(20 < 30 or 30 > 30)
print(20 > 10 or 20 > 10)
print(a != b)
print(not(a != b))
````

* and 연산자는 두 개가 모두 참인 경우에만 True, 그렇지 않으면 False 반환
* or 연산자는 두 개 중 하나 이상이 참인 경우에만 True, 그렇지 않으면 False 반환
* not 연산자는 참과 거짓을 바꾸는 연산자
* 이들을 논리연산자(Logical Operator)이라고 한다

### 17. 1234는 몇 분,몇 초일까
초를 입력하면 '몇 분 몇 초'로 환산해주는 코드를 다음과 같이 출력하도록 작성하세요


````python
n = int(input('초를 입력하세요')) ## input함수는 항상 문자열(str) 을 반환하기 때문에 int로 해줘야함!!
min = n//60 # 정수 나눗셈(목). /는 실수 나눗셈
sec = n%60
print(min,'분',sec,'초')
````

### 18. 7512분은 며칠, 몇 시간, 몇 분일까
분 단위로 입력된 시간을 며칠, 몇 시간, 몇 분으로 환산하여 출력하는 코드를 작성하세요.

````python
n = int(input('분을 입력하세요'))
day = n//(60*24)
hour = (n%(60*24))//60
min = (n%(60*24))%60
print(n,'분은',day,'일',hour,'시간',min,'분')
````

### 24. 공백 리스트와 합치기

````python
a=['우진', '시은']
b=['메이킷', '소피아', '하워드']
c=[]
c.extend(a)
print(c) #[’우진’, ‘시은’]
c.extend(b)
print(c) #[’우진’, ‘시은’, ‘메이킷’, ‘소피아’, ‘하워드’ ]
````

* 리스트와 리스트를 합칠 때는 extend( )함수 사용
* a가 공백 리스트일때 a.extend(b) 는  a=list(b) (리스트 b와 동일한 리스트 복사) 와 동일 결과 나옴

### 26. 리스트 잘라내기(슬라이싱)

다음과 같이 코드 실행 결과가 나오도록 코드의 빈칸을 완성하세요.

````python
a = [1,2, 3, 4, 5, 'a', 'b', 'c', 'd', 'e']

b = a[:5]
print('b :',b)

c = a[-5:]
print('c :',c)

d = a[4:6]
print('d :',d)

e = a[::2]
print('e :',e)

f = a[1::2]
print('f :',f)
````

* a[n1: n2: n3]: 리스트 n1에서부터 n2 직전까지 잘라내되 n3만큼 점프하면서 리스트를 잘라내라

### 29. 리스트 안에 있는 문자로 하나의 문자열 만들기(join)

a = [‘시은’, ’우진’, ’지연’, ’지훈’]과 같이 우리 반 학생들 이름이 저장된 리스트가 있습니다. 이 리스트의 문자열들을 연결(join)해 하나의 문자열을 만들려고 합니다. 다음과 같이 코드 실행 결과가 나오도록 코드의 빈칸을 완성하세요.

````python
a = ['시은', '우진', '지훈', '지연']
b =''.join(a)
print(b)
````

* 문자열로 구성된 리스트에서 문자열을 모두 연결할 때는 join( )을 활용. 문자열변수이름 = ‘문자와문자사이연결문자’.join(리스트)

````python
a = ['시은', '우진', '지훈', '지연']
b =','.join(a)
print(b)
````

### 30. 문자열 분리해 리스트 만들기(split)

a = '시은 우진 지훈 지연'과 같이 우리 반 학생들 이름이 저장된 문자열이 있습니다. 공백이 나올 때마다 문자열을 분리해(split) 하나의 리스트를 만들려고 합니다. 다음과 같이 코드 실행 결과가 나오도록 코드의 빈칸을 완성하세요 .

### 31. 여러 개의 값 입력받기

세 개의 숫자를 입력받아 가장 작은 숫자를 출력하는 코드를 작성하세요.

````python
a,b,c = map(int,input().split())
print(min(a,b,c))
````

* 핵심
  * input( ): 키보드로부터 입력받은 값을 문자열로 전환
  * map( ):  리스트 원소들을 순서대로 일대일로 대응해 변수로 저장
  * min( ): 숫자 중에 가장 작은 수를 구함
